```python
url = "https://learning-analytics.info/journals/index.php/JLA/issue/view/452"
```

```python
respuesta = requests.get(url)
respuesta
```

```python
datos = respuesta.text
datos
```

```python
soup = BeautifulSoup(datos, 'html.parser')
```

```python
soup
```

```python
npo_articulos = {}
articulos_no = 0
```

```python
articulos = soup.find_all('table', {'class':'tocArticle'})
articulos
```

```python
for articulo in articulos:
    titulo = articulo.find('div', {'class':'tocTitle'}).text
    autor = articulo.find('div', {'class':'tocAuthors'}).text
    enlace = articulo.find('a', {'':''}).get('href')
    
    articulos_no +=1
    npo_articulos[articulos_no] = [titulo,autor,enlace]
    
    print('Titulo', titulo, '\nAutor',autor, '\nEnlaces', enlace)
```

```python
print(npo_articulos)
```

```python
npo_articulos_df = pd.DataFrame.from_dict(npo_articulos, orient = 'index', columns = ['Titulo', 'Autores','Enlaces'])
```

```python
npo_articulos_df.head()
```

```python
npo_articulos_df.to_csv('learning_analytics.csv')
```
